```python
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# load data
df = pd.read_csv("../datasets/Training.csv")

# remove unnamed column if exists
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

X = df.drop("prognosis", axis=1)
y = df["prognosis"]

# encode
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# best model
model = RandomForestClassifier(
    n_estimators=300,
    random_state=42
)

model.fit(X, y_encoded)

# save
pickle.dump(model, open("model.pkl","wb"))
pickle.dump(le, open("label_encoder.pkl","wb"))
pickle.dump(X.columns.tolist(), open("columns.pkl","wb"))

print("MODEL TRAINED & SAVED")
import pickle
import numpy as np

def predict_disease(symptoms_list):

    model = pickle.load(open("model.pkl","rb"))
    le = pickle.load(open("label_encoder.pkl","rb"))
    cols = pickle.load(open("columns.pkl","rb"))

    # create input vector
    input_vector = [0]*len(cols)

    for s in symptoms_list:
        s = s.strip().lower()
        if s in cols:
            idx = cols.index(s)
            input_vector[idx] = 1

    input_array = np.array(input_vector).reshape(1,-1)

    # prediction
    probs = model.predict_proba(input_array)[0]

    top3_idx = probs.argsort()[-3:][::-1]

    top3 = []
    for i in top3_idx:
        d = le.inverse_transform([i])[0]
        p = round(probs[i]*100,2)
        top3.append((d,p))

    best_disease = top3[0][0]
    confidence = top3[0][1]

    # 🔥 confidence check like hospital AI
    if confidence < 35:
        return "Not enough symptoms", top3

    return best_disease, top3

print(predict_disease([
'back_pain','weakness_in_limbs','neck_pain','dizziness','muscle_weakness'
]))

```
